```python
import pandas as pd
import re
```

```python
df_poblation = pd.read_csv('resultado/df_poblation.csv')
df_deaths = pd.read_csv('resultado/df.csv')
```

```python
df_deaths.head()
```

```python
df_poblation.head()
```

```python
#Creamos un df de población filtrado a partir de 1990, ya que los datos de muertes por distintas causas empiezan en ese año
df_pob_filtered_year = df_poblation[df_poblation['Year']>=1990]
```

```python
#Hacemos un merge con las columnas Entity y Year

merged_df = pd.merge(df_deaths, df_pob_filtered_year, on=['Entity', 'Year'])
```

```python
merged_df.head()
```

```python
# Identificamos las columnas de muertes (todas excepto 'Entity', 'Code','Year' e 'ISO3')
deaths_columns = [col for col in df_deaths.columns if col not in ['Entity','Year', 'Code', 'ISO3']]
print(deaths_columns)
```

```python
# Aplicamos la fórmula para generar la tasa de muertes por cada 100000k
for col in deaths_columns:
   new_column = f'{col}_per_100k',
   merged_df[new_column] = merged_df.apply(
        lambda row: (row[col] / row['Population']) * 100000, axis=1
    )
```

```python
merged_df.head()
```

```python
# Eliminamos las columnas originales de muertes
merged_df.drop(columns=deaths_columns, inplace=True)
```
